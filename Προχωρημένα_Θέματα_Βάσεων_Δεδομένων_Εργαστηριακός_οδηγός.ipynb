### 0. Προκαταρκιτκά - Διαμόρφωση περιβάλλοντος εργασίας Εικονικών Μηχανών στην πλατφόρμα okeanos-knossos

Στα προαπαιτούμενα για την εξαμηνιαία εργασία του μαθήματος είναι η δημιουργία Εικονικών Μηχανών (Virtual Machines - VMs) και η διαμόρφωση του περιβάλλοντος εργασίας με τα frameworks Apache Hadoop και Apache Spark. Ακολουθεί αναλυτικός οδηγός για τους σπουδαστές σχετικά με τη διαμόρφωση του συγκεκριμένου περιβάλλοντος.

#### 0.1 Δημιουργία λογαριασμού στην υπηρεσία cloud *~okeanos-knossos*
To *~okeanos-knossos* είναι μια IAAS (Infrastructure-as-a-Service) υπηρεσία που επιτρέπει στους χρήστες του να δημιουργούν και να διαχειρίζονται εικονικά υπολογιστικά συστήματα. Τα τελευταία χρόνια, τέτοιου είδους υπηρεσίες όπως το Amazon AWS, Google Cloud, Microsoft Azure έχουν κυριαρχήσει στην αγορά, καθώς δίνουν τη δυνατότητα στους πελάτες τους να έχουν πρόσβαση σε υπολογιστικούς πόρους χωρίς να επωμίζονται το κόστος συντήρησης της υλικής υποδομής. Σε αντίθεση με τις εμπορικές αυτές υπηρεσίες, το *~okeanos-knossos* έχει αναπτυχθεί για ερευνητικούς σκοπούς και παρέχεται στους σπουδαστές δωρεάν.

Στα πλαίσια της παρούσας εργασίας ζητείται από τους σπουδαστές αρχικά να δημιουργήσουν έναν λογαριασμό στην υπηρεσία *~okeanos-knossos*, ακολουθώντας το σύνδεσμο: https://okeanos-knossos.grnet.gr/home/.

Επιλέγοντας την εγγραφή χρησιμοποιώντας τον ακαδημαϊκό λογαριασμό του ΕΜΠ, οι σπουδαστές μπορούν να ολοκληρώσουν τη δημιουργία του ακαδημαϊκού λογαριασμού τους τους στην υπηρεσία.


#### 0.2 Εγγραφή στο *project* του μαθήματος
Στη συνέχεια, προκειμένου να αποδοθούν οι υπολογιστικοί πόροι που απαιτούνται, οι σπουδαστές θα πρέπει εγγραφούν στο *project* που αντιστοιχεί στο μάθημα των **Προχωρημένων Θεμάτων Βάσεων Δεδομένων**. Συγκεκριμένα, ακολουθόντας τον σύνδεσμο: https://astakos.okeanos-knossos.grnet.gr/ui/projects/search, οι σπουδαστές θα πρέπει να αναζητήσουν το *project* `advancedDB.dblab.ntua.gr` και να δημιουργήσουν ένα αίτημα εγγραφής.

#### 0.3 Δημιουργια ζεύγους κρυπτογραφικών κλειδιών και καταχώρισή του στην υπηρεσία *~okeanos-knossos*
Προτού δοθούν πηροφορίες σχετικές με τη δημιουργία των Εικονικών Μηχανών, κρίνεται σκόπιμο να εξασφαλιστεί η ύπαρξη ζεύγους κρυπτογραφικών κλειδιών στον ηλεκτρονικό υπολογιστή από τον οπoίο οι χρήστες θα έχουν απομακρυσμένη πρόσβαση προς τους πόρους τους στην υπηρεσία  *~okeanos-knossos*. Προκειμένου να δημιουργηθεί ένα νέο ζεύγος κρυπτογραφικών κλειδιών, αρκεί να εκτελέστεί, (σε περιβάλλον linux), η παρακάτω εντολή και να ακολουθηθούν οι οδηγίες:`$ ssh-keygen`.

- Εναλλακτικά, ένας πλήρης οδηγός για τη διαδικασία δημιουργίας ζεύγους κλειδιών υπάρχει στο σύνδεσμο: https://www.digitalocean.com/community/tutorials/how-to-set-up-ssh-keys-on-ubuntu-20-04.

Με την ολοκλήρωση της διαδικασίας, τα περιεχόμενα του δημοσίου κλειδιού θα εμφανιστούν στην οθόνη με την εντολή:
```
$ cat ~/.ssh/id_rsa.
```
Στη συνέχεια, θα πρέπει να γίνει η καταχώριση του δημόσιου κλειδιού  στο λογαριασμό της υπηρεσίας *~okeanos-knossos*. Ακολουθώντας το σύνδεσμο https://cyclades.okeanos-knossos.grnet.gr/ui/#public-keys/ και επιλέγοντας τη δημιουργία νέου ζεύγους κλειδιών, οι χρήστες μπορούν να εισάγουν το κλειδί τους και να το καταχωρίσουν στο λογαριασμό τους με ένα όνομα που θα επιλέξουν.

#### 0.4 Δημιουργία Εικονικών Μηχανών
Το επόμενο βήμα είναι η δημιουργία των Εικονικών Μηχανών (VMs) στις οποίες θα πραγματοποιηθεί η ανάπτυξη του ζητούμενου λογισμικού καθώς και η εξέταση της εργασίας. Στο σύνδεσμο https://cyclades.okeanos-knossos.grnet.gr/ui/#machines/icon/, οι χρήστες έχουν τη δυνατότητα δημιουργίας νέων VMs. Για τις ανάγκες της παρούσας εργασίας θα χρειαστούμε 2 VMs με τα παρακάτω χαρακτηριστικά:
 - Ubuntu Server LTS 16.04 OS
 - 4 CPUs
 - 8GB RAM
 - 30GB disk capacity

Αφού ξεκινήσουν τη δημιουργία νέου VM, οι σπουδαστές θα πρέπει να κάνουν τις παραπάνω επιλογές στα μενού που ακολουθούν.

**Προσοχή!**
 - Είναι σημαντικό κατά τη διαδικασία δημιουργίας των Εικονικών Μηχανών να επλιχθεί το δημόσιο κλειδί που δημιουργήθηκε στο προηγούμενο βήμα, αλλιώς δε θα είναι δυνατή η πρόσβαση στα VMs.
 - Αφού δημιουργηθούν τα VMs είναι σημαντικό να αποθηκευτεί το *password* που το σύστημα δημιουργεί αυτόματα για το χρήστη, διότι σε περίπτωση απώλειας η διαδικασία θα πρέπει να επαναληφθεί.


Εφόσον οι εικονικές μηχανές δημιουργηθούν επιτυχώς, οι χρήστες μπορούν να αποκτήσουν πρόσβαση σε αυτές με την εντολή:
```
$ ssh user@snf-****-ok-kno.grnetcloud.net
```

Επειδή πρόκειται για την πρώτη φορά που ζητείται απομακρυσμένη σύνδεση στο συγκεκριμένο σύστημα, θα ζητηθεί από το τοπικό λειτουργικό σύστημα επιβεβαίωση της ενέργειας:
```
Are you sure you want to continue connecting (yes/no/[fingerprint])? yes
```

Στη συνέχεια, προτείνεται να γίνει αλλαγή του αυτόματα δημιουργημένου password με την εντολή:
```
$ passwd
```

Τα ίδια βήματα θα πρέπει να ακολουθηθούν και για τη δημιουργία του δεύτερου VM.

#### 0.5 Διαμόρφωση public και private network
Η υπηρεσία *~okeanos-knossos* μας παρέχει μια μοναδική δημόσια ipv4. Η διαχείριση των πόρων δικτύου γίνεται από το σύνδεσμο: https://cyclades.okeanos-knossos.grnet.gr/ui/#networks/. Προτείνεται οι σπουδαστές να αποδώσουν τη συγκεκριμένη ip σε ένα από τα VMs τους (master). Στη συνέχεια, σε βήματα που θα ακολουθήσουν θα χρειαστεί η δημόσια ip να αποσυνδεθεί από το συγκεκριμένο VM και να συνδεθεί στο άλλο - για να ολοκληρωθεί επιτυχώς η διαδικασία αναβάθμισης της αρχικής έκδοσης του Λειτουργικού Συστήματος (βλέπε 1.2, παρακάτω).

- **Να είστε πολύ προσεκτικοί με τη χρήση της *public* ip. Μπορείτε να χρησιμοποιήσετε κάποιο firewall, να αποσυνδέετε την ip όταν δεν τη χρησιμοποιείτε ή και να απενεργοποιείτε εντελώς τις εικονικές σας μηχανές. Σε περίπτωση επίθεσης υπάρχει περίπτωση η ip σας να χαρακτηριστεί αυτόματα ως ευάλωττη και να χάσετε την πρόσβαση στο ipv4 address space - και φυσικά την πρόσβαση στις web εφαρμογές του cluster σας!**

Στη συνέχεια, προς διευκόλυνσή, προτείνεται η αλλαγή των hostnames των VMs. Εκτελούνται στο master και το worker αντίστοιχα:

```bash
$ sudo hostnamectl set-hostname okeanos-master
$ sudo hostnamectl set-hostname okeanos-worker
```

Ακολουθεί η διαμόρφωση του ιδιωτικού δικτύου μέσω του οποίου θα επικοινωνούν τα δύο VMs. Ακολουθώντας το σύνδεσμο: https://cyclades.okeanos-knossos.grnet.gr/ui/#networks/, οι χρήστες θα πρέπει να δημιουργήσουν ένα ιδιωτικό ipv4 δίκτυο και να αποδώσουν μια ip που ανήκει σε αυτό σε κάθε VM.

Κατόπιν, καταχωρούμε την ip του ιδιωτικού δικτύου μας και στα δύο μηχανήματα με το αντίστοιχο όνομα στο αρχείο `/etc/hosts`.
Η τελική εικόνα του συγκεκριμένου αρχείου θα πρέπει να είναι η παρακάτω και για τους δύο κόμβους:
```text
127.0.0.1       localhost
192.168.0.xxx   okeanos-master
192.168.0.xxx   okeanos-worker


# The following lines are desirable for IPv6 capable hosts
::1             localhost ip6-localhost ip6-loopback
ff02::1         ip6-allnodes
ff02::2         ip6-allrouters
```
Αποθηκεύουμε και επανεκκινούμε τα VMs με την εντολή:
```
$ sudo reboot
```

Αφού τα VMs επανεκκινήσουν, συνδεόμαστε στο master και δημιουργούμε ένα νέο κρυπτογραφικό κλειδί το οποίο θα χρησιμοποιηθεί ώστε αυτός ο κόμβος να έχει πρόσβαση τόσο στον εαυτό του όσο και στον worker μέσω `ssh`. Αντίστοιχα με τα προηγούμενα (βήμα 0.3):
```bash
$ ssh-keygen -t rsa #press [ENTER] as many times as promted
$ cat ~/.ssh/id_rsa.pub >> ~/.ssh/authorized_keys
```
Τυπώνουμε τα περιεχόμενα του κλειδιού στο terminal με την παρακάτω εντολή και το εισάγουμε στο worker VM:
```bash
$ cat ~/.ssh/id_rsa.pub #master
$ vim ~/.ssh/authorized_keys #worker: copy above output and paste it here
```
Επιβεβαιώνουμε την πρόσβαση με ssh και στα δύο VMs.
```bash
$ ssh okeanos-master
$ ssh okeanos-worker
```

### 2. Eγκατάσταση και διαμόρφωση Apache Spark over YARN στο μικρό cluster Εικονικών Μηχανών
**Σημείωση:** Οι παρακάτω οδηγίες θεωρούν ως δεδομένο ότι οι ενέργειες που περιγράφονται παραπάνω στον ίδιο οδηγό έχουν ακολουθηθεί πιστά. Οι σπουδαστές είναι ελεύθεροι να αλλάξουν τα δεδομένα κομμάτια κώδικα ανάλογα με τις δικές τους προτιμήσεις στο configuration.

#### 2.1 Εγκατάσταση java
Συνεχίζουμε με την εγκατάσταση του Apache Hadoop στο νέο μας cluster. Αρχικά θα πρέπει να εγκατασταθεί σε όλους τους κόμβους η έκδοση java που υποστηρίζεται από τα επίσημα αποθετήρια του λειτουργικού μας συστήματος: `$ sudo apt install default-jdk -y`
Η επιτυχής εγκατάσταση μπορεί να επαληθευτεί με την εντολή: `$ java -version`.

#### 2.2 Εγκατάσταση Hadoop και Apache Spark
Προκειμένου να εγκαταστήσουμε τις τελευταίες εκδόσεις των Apache Spark και Hadoop, δημιουργούμε το directory `~/opt` όπου αποθηκεύουμε τα εκτελέσιμα των τελευταίων εκδόσεων των δύο Apache projects.
```bash
$ mkdir ./opt
$ mkdir ./opt/bin
```
Ακολούθως, κατεβάζουμε τα συμπιεσμένα αρχεία από τα επίσημα websites των δύο projects, αποσυμπιέζουμε τα περιεχόμενα, τα μεταφέρουμε στο directory `~/opt/bin` και δημιουργούμε links στο κύριο directory `~/opt`.
```bash
$ wget https://dlcdn.apache.org/hadoop/common/hadoop-3.3.6/hadoop-3.3.6.tar.gz
$ tar -xvzf hadoop-3.3.6.tar.gz
$ mv hadoop-3.3.6 ./opt/bin
$ wget https://dlcdn.apache.org/spark/spark-3.5.0/spark-3.5.0-bin-hadoop3.tgz
$ tar -xvzf spark-3.5.0-bin-hadoop3.tgz
$ mv ./spark-3.5.0-bin-hadoop3 ./opt/bin/
$ cd ./opt
$ ln -s ./bin/hadoop-3.3.6/ ./hadoop
$ ln -s ./bin/spark-3.5.0-bin-hadoop3/ ./spark
$ cd
$ rm hadoop-3.3.6.tar.gz
$ rm spark-3.5.0-bin-hadoop3.tgz
$ mkdir ~/opt/data
$ mkdir ~/opt/data/hadoop
$ mkdir ~/opt/data/hdfs
```

#### 2.3 Διαμόρφωση μεταβλητών περιβάλλοντος (environment variables)
Αρχικά, επεξεργαζόμαστε με πρόγραμμα της αρεσκείας μας το αρχείο `~/.bashrc` το οποίο εκτελείται κάθε φορά που ανοίγει κάποια σύνδεση terminal προς τις εικονικές μας μηχανές, προσθέτωντας τις παρακάτω γραμμές:
```bash
export JAVA_HOME=/usr/lib/jvm/java-11-openjdk-amd64  #Value should match: dirname $(dirname $(readlink -f $(which java)))
export HADOOP_HOME=/home/user/opt/hadoop
export SPARK_HOME=/home/user/opt/spark
export HADOOP_INSTALL=$HADOOP_HOME
export HADOOP_MAPRED_HOME=$HADOOP_HOME
export HADOOP_COMMON_HOME=$HADOOP_HOME
export HADOOP_HDFS_HOME=$HADOOP_HOME
export HADOOP_YARN_HOME=$HADOOP_HOME
export HADOOP_COMMON_LIB_NATIVE_DIR=$HADOOP_HOME/lib/native
export PATH=$PATH:$HADOOP_HOME/sbin:$HADOOP_HOME/bin:$SPARK_HOME/bin;
export HADOOP_CONF_DIR=$HADOOP_HOME/etc/hadoop
export HADOOP_OPTS="-Djava.library.path=$HADOOP_HOME/lib/native"
export LD_LIBRARY_PATH=/home/ubuntu/opt/hadoop/lib/native:$LD_LIBRARY_PATH
export PYSPARK_PYTHON=python3
```
Ενημερώνουμε το περιβάλλον εκτελώντας το script με τις αλλαγές που εφαρμόσαμε:
```bash
$ source ~/.bashrc
```
#### 2.4 Διαμόρφωση περιβάλλοντος για το Hadoop Distributed File System
Τα configuration files του Hadoop βρίσκονται στο φάκελο `/home/user/opt/hadoop/etc/hadoop`. Έχοντας ήδη ορίσει τη μεταβλητή `HADOOP_HOME` να δείχνει στο directory `/home/user/opt/hadoop`, αναφερόμαστε στο ίδιο path γράφοντας: `$HADOOP_HOME/etc/hadoop/`. Παρακάτω καταγράφουμε τις αλλαγές που πρέπει να γίνουν:

Ξεκινάμε με το αρχείο `$HADOOP_HOME/etc/hadoop/hadoop-env.sh`, όπου πρέπει να προστεθεί η γραμμή:
```bash
export JAVA_HOME=/usr/lib/jvm/java-11-openjdk-amd64
```

Διαμορφώνουμε το αρχείο `$HADOOP_HOME/etc/hadoop/core-site.xml` ώστε τα περιεχόμενά του να είναι τα ακόλουθα:
```xml
<?xml version="1.0" encoding="UTF-8"?>
<?xml-stylesheet type="text/xsl" href="configuration.xsl"?>
<configuration>
    <property>
        <name>hadoop.tmp.dir</name>
        <value>/home/user/opt/data/hadoop</value>
        <description>Parent directory for other temporary directories.</description>
    </property>
    <property>
        <name>fs.defaultFS </name>
        <value>hdfs://okeanos-master:54310</value>
        <description>The name of the default file system. </description>
    </property>
</configuration>
```
Επίσης, διαμορφώνουμε το αρχείο `$HADOOP_HOME/etc/hadoop/hdfs-site.xml` ώστε τα περιεχόμενά του να είναι τα ακόλουθα:
```xml
<?xml version="1.0" encoding="UTF-8"?>
<?xml-stylesheet type="text/xsl" href="configuration.xsl"?>
<configuration>
    <property>
        <name>dfs.replication</name>
        <value>1</value>
        <description>Default block replication.</description>
    </property>
    <property>
        <name>dfs.datanode.data.dir</name>
        <value>/home/user/opt/data/hdfs</value>
    </property>
</configuration>
```
 Τέλος δημιουργούμε το αρχείο `$HADOOP_HOME/etc/hadoop/workers` με τα παρακάτω περιεχόμενα:
 ```text
 okeanos-master
 okeanos-worker
 ```
 Εγγράφουμε και τους δύο μας κόμβους, καθώς ο master θα εκτελέσει και χρέη worker ώστε να αναδειχθεί η κατανομή των εργασιών στο μικρό μας cluster.

#### 2.6 Διαμόρφωση περιβάλλοντος για το Hadoop YARN
Επεξεργαζόμαστε το αρχείο `$HADOOP_HOME/etc/hadoop/yarn-site.xml`, η τελική μορφή του οποίου πρέπει να είναι η εξής:

```xml
<?xml version="1.0"?>
<configuration>
<!-- Site specific YARN configuration properties -->
    <property>
        <name>yarn.resourcemanager.hostname</name>
        <value>okeanos-master</value>
    </property>
    <property>
        <name>yarn.resourcemanager.webapp.address</name>
        <!--Insert the public IP of your master machine here-->
        <value>83.212.xxx.xxx:8088</value>
    </property>
    <property>
        <name>yarn.nodemanager.resource.memory-mb</name>
        <value>6144</value>
    </property>
    <property>
        <name>yarn.scheduler.maximum-allocation-mb</name>
        <value>6144</value>
    </property>
    <property>
        <name>yarn.scheduler.minimum-allocation-mb</name>
        <value>128</value>
    </property>
    <property>
        <name>yarn.nodemanager.vmem-check-enabled</name>
        <value>false</value>
    </property>
   <property>
        <name>yarn.nodemanager.aux-services</name>
        <value>mapreduce_shuffle,spark_shuffle</value>
    </property>
    <property>
        <name>yarn.nodemanager.aux-services.mapreduce_shuffle.class</name>
        <value>org.apache.hadoop.mapred.ShuffleHandler</value>
    </property>
    <property>
        <name>yarn.nodemanager.aux-services.spark_shuffle.class</name>
        <value>org.apache.spark.network.yarn.YarnShuffleService</value>
    </property>
    <property>
        <name>yarn.nodemanager.aux-services.spark_shuffle.classpath</name>
        <value>/home/user/opt/spark/yarn/*</value>
    </property>
</configuration>

```
Εκκινούμε το YARN με την εντολή:
 ```bash
 $ start-yarn.sh
 ```
Επιβεβαιώνουμε ότι η διαδικασία έχει επιτύχει προσπελαύνοντας με ένα browser το  web interface στην public ip του master: `http://83.212.xxx.xxx::8088/cluster`. Θα πρέπει να βλέπουμε 2 διαθέσιμους live nodes.

```bash
$ hadoop fs -mkdir /spark.eventLog
$ $SPARK_HOME/sbin/start-history-server.sh
```
